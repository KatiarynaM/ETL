Добавьте графики с досрочным погашением по этим параметрам. Т.е. линии по выплатам основного долга и процентов если платеж будет 120 или 150 тыс. руб. В результате должно получиться 6 линий. Используйте разные цвета.

```python

df1 = pd.read_excel('s4_2.xlsx', sheet_name='Лист1')

df2 = pd.read_excel('s4_2.xlsx', sheet_name='120 тыс.')

df3 = pd.read_excel('s4_2.xlsx', sheet_name='150 тыс.')
```

```python
x1 = df1['Месяц']
y1 = df1['Платеж по основному долгу']
y2 = df1['Платеж по процентам']
x2 = df2['Месяц']
y3 = df2['Платеж по основному долгу']
y4 = df2['Платеж по процентам']
x3 = df3['Месяц']
y5 = df3['Платеж по основному долгу']
y6 = df3['Платеж по процентам']

plt.title('График платежей')
plt.xlabel('Сумма')
plt.ylabel('Месяц')

plt.plot(x1, y1, 'b')    # синяя линия
plt.plot(x1, y2, '--b')
plt.plot(x2, y3, 'r')
plt.plot(x2, y4, '--r')
plt.plot(x3, y5, 'g')
plt.plot(x3, y6, '--g')


plt.show()
```
