## Напишите функцию, которая добавит колонку с разбиением возраста по группам: до 20, от 20 до 30, от 30 до 36 и старше 36

```python
df ['age_group'] = 0

group = [(df['Age'] < 20),(df['Age'] >= 20) & (df['Age'] < 36), (df['Age'] >= 36)]
values = ['under 20', 'between', 'over 36']

df['age_group']= np.select(group, values)

df.head(10)
```
